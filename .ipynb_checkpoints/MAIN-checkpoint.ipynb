```python
##################### Все константы
################################################

MAX_LEN_OF_SMALL_WORD = 3 # слова длиной <= числа будут удалены

################################################
```

```python
%%time
# ПОЛУЧЕНИЕ ВСЕХ ОБУЧАЮЩИХ ТЕКСТОВ

MAIN_LIST_ALL_THEMES = []


list_path = ["THEMES\ЯД_ФИЗ" , "THEMES\AI"]


for path in list_path:
    #LIST_TEXTS_THEME = 
    MAIN_LIST_ALL_THEMES.append( GET_LIST_TEXTS_FROM_THEME( path ) ) #, "PRINT_INFO")



#print(MAIN_LIST_ALL_THEMES[1][0])
```

```python
%%time
# Уборка плохих символов
MIAN_LIST_THEMES_FINAL_TEXTS = []

for list_theme_texts in MAIN_LIST_ALL_THEMES:
    #LIST_FINAL_TEXTS =  
    MIAN_LIST_THEMES_FINAL_TEXTS.append( GET_scrubbed_list( list_theme_texts ) )


#print(MIAN_LIST_THEMES_FINAL_TEXTS[1][0])
```

```python
%%time
# ДЕЛИМ НА СЛОВА
MAIN_LIST_LIST_TEXT_WORDS = []


for all_one_theme_texts in MIAN_LIST_THEMES_FINAL_TEXTS:
    
    LIST_TEXTS_WORDS_one_theme = [] # Каждую тему новый
    for one_text in all_one_theme_texts:
        
        list_words = one_text.split(" ")
        LIST_TEXTS_WORDS_one_theme.append( GET_LIST_WORDS_NO_SMALL( list_words , MAX_LEN_OF_SMALL_WORD ) )
    
    
    
    MAIN_LIST_LIST_TEXT_WORDS.append( LIST_TEXTS_WORDS_one_theme )
        
        
        
#print(MAIN_LIST_LIST_TEXT_WORDS[1][0])
```

```python
%%time
########## Обрезка и графики

#### Смещения
offset_min = 1.0   # 0 + СКО * это
offset_max = 1.0   # max - СКО * это
offset_M = 1.0   # по обе стороны
#draw_plt = False
draw_plt = True
####

MAIN_THEMES_TEXTS_WORDS_EDGED = []


for all_one_theme_texts in MAIN_LIST_LIST_TEXT_WORDS:
    print("#"*40)
    buf_list = []
    for one_word_set in all_one_theme_texts:
        buf_list.append(  GET_EDGED_LIST_WORDS_SKO_PLT( one_word_set , [offset_min,offset_max,offset_M] , draw_plt )  )
    
    MAIN_THEMES_TEXTS_WORDS_EDGED.append( buf_list )

```

```python
%%time
# Сшиваем листы      1 тема=1 текст    list[0]=итоговые слова по 0 теме

FINAL_ONE_THEME_ONE_TEXT = []

for all_one_theme_texts in MAIN_THEMES_TEXTS_WORDS_EDGED:
    buf_list_all = []
    for one_word in all_one_theme_texts:
        if one_word not in buf_list_all:
            buf_list_all.extend( one_word )
        
    FINAL_ONE_THEME_ONE_TEXT.append( buf_list_all )

    
    
```
